```python
import matplotlib.pyplot as plt
import numpy as np
```

```python


def plane(x, y, z):
    return (z + x**2 + y**2)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

# Values of z to visualize
z_values = [-10, 0, 10, 20]  # Change these as needed

# Set up the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the planes for each z value
for z in z_values:
    z_plane = plane(x, y, z)
    ax.plot_surface(x, y, z_plane, alpha=0.5, label=f'z = {z}')

# Add labels and a legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title("Planes for X^2 + Y^2 = z")
plt.show()
```

```python
def line(x, z):
    return (z - 2 * x) / 3

# Generate x values
x = np.linspace(-10, 10, 500)

# Different z values
z_values = [-10, 0, 10, 20]

# Plot each line
plt.figure(figsize=(8, 6))
for z in z_values:
    y = line(x, z)
    plt.plot(x, y, label=f'z = {z}')

# Add labels, legend, and grid
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Lines for 2x + 3y = z')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
